```python
Purchase_Summary = pd.read_sql(
    """
    SELECT
    p.Brand,
    p.VendorNumber,
    p.VendorName,
    SUM(p.Quantity) AS Total_Qantity_Purchase,
    SUM(p.Dollars) AS Total_Purchase_Dollar,
    pp.Description,
    pp.Price AS Actual_Price_Of_Product,
    pp.PurchasePrice AS Actual_Purchase_Price,
    pp.Volume
    FROM purchases AS p
    JOIN purchase_prices AS pp
    ON p.Brand = pp.Brand
    WHERE pp.PurchasePrice > 0
    GROUP BY 
        p.Brand,
        p.VendorNumber,
        p.VendorName,
        pp.Description,
        pp.Price,
        pp.PurchasePrice,
        pp.Volume
    ORDER BY Actual_Purchase_Price
    """,
    Conn
)
```

```python
Purchase_Summary
```

```python
Freight_Summary = pd.read_sql(
    """
    SELECT
        VendorNumber,
        VendorName,
        SUM(Freight) AS Total_Freight_Cost
        FROM vendor_invoice
        GROUP BY
            VendorNumber,
            VendorName
    """,
    Conn
)
Freight_Summary
```

```python
Sale_Summary = pd.read_sql(
    """
    SELECT
        Brand,
        VendorNo,
        VendorName,
        Description,
        SUM(SalesPrice) AS Total_Sale_Price ,
        SUM(SalesQuantity) AS Total_Sales_Quantiy,
        SUM(SalesDollars) AS Total_Sale_Dollar,
        SUM(ExciseTax) AS Total_Excise_Tax,
        Volume
        FROM sales
        GROUP BY
            Brand,
            VendorNo,
            VendorName,
            Description,
            Volume
    """,
    Conn
)
```

```python
Sale_Summary = Sale_Summary.rename(columns={'VendorNo':'VendorNumber'})
```

```python
Purchase_Summary['Volume'] = Purchase_Summary['Volume'].astype('float64')
```

```python
Sale_Summary
```

```python
Vendor_Sale_Summary = pd.merge(
    Purchase_Summary,
    Sale_Summary,
    on = ['Brand','VendorNumber','VendorName','Description','Volume'],
    how = 'left'
)
```

```python
Vendor_Sale_Summary = pd.merge(
    Vendor_Sale_Summary,
    Freight_Summary,
    on = ['VendorNumber','VendorName'],
    how = 'left'
)
```

```python
Vendor_Sale_Summary['Gross_Profit'] = Vendor_Sale_Summary['Total_Sale_Dollar'] - Vendor_Sale_Summary['Total_Purchase_Dollar']
```

```python
Vendor_Sale_Summary['Profit_Margin'] = (Vendor_Sale_Summary['Gross_Profit']/Vendor_Sale_Summary['Total_Sale_Dollar']) * 100
```

```python
Vendor_Sale_Summary['Stock_Turnover'] = Vendor_Sale_Summary['Total_Sales_Quantiy'] / Vendor_Sale_Summary['Total_Qantity_Purchase']
```

```python
Vendor_Sale_Summary['Sales_Pruchase_Ratio'] = Vendor_Sale_Summary['Total_Sale_Dollar']/Vendor_Sale_Summary['Total_Purchase_Dollar']
```
